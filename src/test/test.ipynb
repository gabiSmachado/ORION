```python
import pandas as pd
import requests
import os
import json
import time
```

```python
df = pd.read_csv('intent_dataset.csv',delimiter=";")
```

```python
for index,row in df.iterrows():
    url = "http://mcp-client.smo:9100/intent"
    payload = {
        "intent": (
            f"{row.intent}"
        )
    }
    headers = {"Content-Type": "application/json"}

    response = requests.post(url, json=payload, headers=headers)

    time.sleep(60)
```
